```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/chain-dataset-All-Treated-.csv')
```

```python
df
```

## Unsupervised Learning Approach
- Given the data, we want to correlate the timeframes with the amount of liquidity present in the market.
- These relation can be made with Algorithms that relate the data with similiar caracteristics. Normaly this tools are present in the Unsupervised Learning field.

```python
# Uniform data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cols = np.array(df.columns)
cols = cols[1:]

df_scaled = pd.DataFrame(scaler.fit_transform(df[cols]), columns=cols)
```

```python
df_scaled
```

## Validation && Model

Since the dataset is quite small, Leave-One-Out Cross Validation should be used to garantee no overfitting, however since its unsupervised learning is not quite simple. A better approach would be using a silhouette score to evaluate the performance of the K-means algorithm. The silhouette score is a measure of how similar an object is to its own cluster compared to other clusters.

```python
from sklearn.cluster import KMeans

X = df_scaled[cols]

model = KMeans(n_clusters=10, random_state=42, n_init="auto")
x_kmeans = model.fit(X)
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X, x_kmeans.labels_)
score
```

```python
# Adding cluster information from KMeans to the dataframe
df_scaled['cluster'] = model.labels_

# Grouping dataset by cluster and calculating total liquidity for each protocol over time (mean value)
grouped = df_scaled.groupby('cluster').mean()
```

```python
# grouped.insert(0, column='Protocol', value=df['Protocol'])
grouped
```

```python
df_viz = grouped.transpose()
df_viz = df_viz.reset_index()
df_viz = df_viz.rename(columns = {'index':'date'})
df_viz['date'] = pd.to_datetime(df_viz['date'], format = '%d/%m/%Y')
df_viz = df_viz.set_index('date')

fig, axs = plt.subplots(nrows=1, figsize=(14, 14))
fig.suptitle('Liquidity over Time for All Clusters')

colors = ['blue', 'orange', 'green', 'red', 'magenta', 'grey', 'black', 'cyan', 'brown', 'violet']

for i in range(2,10):
    label = df['Protocol'][i]
    axs.plot(df_viz.index, df_viz[i], linestyle='solid', marker=None, color = colors[i], label=label)
axs.legend()

plt.xlabel('Dates')
plt.show()
```
